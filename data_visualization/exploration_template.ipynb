```python
# Let´s plot some barcharts
base_color = sb.color_palette()[0]

# Number of books at home
sb.countplot(data=df_clean, 
             x='how_many_books_at_home', 
             color=base_color, 
             order=df_clean['how_many_books_at_home'].value_counts().index);
plt.xticks(rotation=90)

df_clean['how_many_books_at_home'].value_counts()

# Add function to add annotations
def add_percentage_annotations(df, column):
    n_points = df.shape[0]
    cat_counts = df[column].value_counts()
    locs, labels = plt.xticks() # get the current tick locations and labels

    # loop through each pair of locations and labels
    for loc, label in zip(locs, labels):

        # get the text property for the label to get the correct count
        count = cat_counts[label.get_text()]
        pct_string = '{:0.1f}%'.format(100*count/n_points)

        # print the annotation just below the top of the bar
        plt.text(loc, count-8, pct_string, ha = 'center', color = '0')
        
add_percentage_annotations(df_clean, 'how_many_books_at_home')
```

```python
# Gender
sb.countplot(data=df_clean, x='gender', color=base_color)
add_percentage_annotations(df_clean, 'gender')
```

```python
sb.countplot(data=df_clean, 
             x='immigration_status',
             color=base_color,
             order=df_clean['immigration_status'].value_counts().index);

add_percentage_annotations(df_clean, 'immigration_status')
```

```python
sb.countplot(data=df_clean, 
             x='at_school_-_portable_laptop',
             color=base_color,
             order=df_clean['at_school_-_portable_laptop'].value_counts().index);

add_percentage_annotations(df_clean, 'at_school_-_portable_laptop')
```
